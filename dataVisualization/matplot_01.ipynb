```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Figure 
畫布, 上方可有多個子圖

## Axes
子圖, 每個子圖一次只能畫在一張圖上

```python
fig_01 = plt.figure(figsize=(10,6))

ax_01 = fig_01.add_subplot(2, 2, 1) #將 ax_01 放在將原圖分割成2*2中 編號1的位置 -> (221)
ax_02 = fig_01.add_subplot(2, 2, 2)
ax_03 = fig_01.add_subplot(2, 2, 3)
ax_04 = fig_01.add_subplot(2, 2, 4)
plt.plot([1,2,3], [2,1,7], color='r')          # plt會直接找最近的子圖來繪圖
ax_01.plot([3,2,1], [5,7,1], color='g', linestyle='--')   # 直接用指定子圖來繪製
```

## 快速生成大量子圖

```python

fig_02, axes = plt.subplots(3, 3, sharex=True, sharey=False)
'''        
sharex, sharey : bool or {'none', 'all', 'row', 'col'}, default: False
Controls sharing of properties among x (*sharex*) or y (*sharey*)
axes:

- True or 'all': x- or y-axis will be shared among all subplots.
- False or 'none': each subplot x- or y-axis will be independent.
- 'row': each subplot row will share an x- or y-axis.
- 'col': each subplot column will share an x- or y-axis.
'''
data = pd.read_csv('Salary_data.csv')

axes[2, 1].scatter(x=data['YearsExperience'], y=data['Salary'], linestyle='--', color='g')
axes[1,1].plot(data, 'g--')
# axes[2, 1].set_facecolor('r')
```

```python
plt.rc('figure', figsize=(10,10))

plt.plot(data['YearsExperience'], data['Salary'], 'g--', marker='v', mfc='r', mec='b')

```
