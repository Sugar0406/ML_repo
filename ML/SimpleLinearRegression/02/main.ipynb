```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("Customers.csv")
data.info()
data
```

```python
IndependentV = data["Annual Income ($)"].to_numpy()

DependentV = data["Spending Score (1-100)"].to_numpy()

x_train, x_test, y_train, y_test = train_test_split(IndependentV, DependentV, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.reshape(-1, 1), y_train.reshape(-1, 1))
predict = model.predict(x_test.reshape(-1, 1)).reshape(-1,)
for i in range(0, len(predict)):
    print((x_test[i],predict[i]))
```

```python
fig = plt.figure(figsize=[15,15])
ax = fig.add_subplot(1,1, 1)
ax.scatter(x_train, y_train, color="#4DE680", label="train_set")
ax.scatter(x_test, y_test, color='r', label="test_set")
ax.plot(x_test, predict, label="regression_line")
ax.legend(fontsize=14)
ax.set_title("The Simple Regression Use Annual Income", fontsize=20)
```
